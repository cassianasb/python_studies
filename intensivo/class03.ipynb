```python
# 1 Entender o desafio
# 2 Entender a area/empresa
# 3 Extracao/Obtencao de dados
# 4 Ajuste de Dados (Tratamento/Limpeza)
# 5 Analise Exploratoria
# 6 Modelagem + Algoritmos
# 7 Interpretacao de Dados

# 1 Importar Base de Dados
import pandas as pd
ads_table = pd.read_csv((r"C:\Users\kassi\Downloads\Vendas - Dez.xlsx")
display(ads_table)
ads_table.info()
# 2 Analise de Exploratoria
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(ads_table)
plt.show()

sns.heatmap(ads_table.corr(), cmap="Wistia", annot=True)
plt.show()

# 3 Modelagem + Algoritmos
from sklearn.model_selection import train_test_split

x = ads_table.drop("Vendas", axis=1)
y = ads_table["Vendas"]

x_test, x_prod, y_test, y_prod = train_test_split(x, y, test_size=0.3)

# 4 Qual Modelo
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

linear = LinearRegression()
randomForest = RandomForestRegressor()

linear.fit(x_prod, y_prod)
randomForest.fit(x_prod, y_prod)

from sklearn import metrics

linear_test = linear.predict(x_prod)
random_test = randomForest.predict(x_prod)

linear_r2 = metrics.r2_score(y_prod, linear_prod)
random_r2 = metrics.r2_score(y_prod, random_prod)
print(linear_r2, random_r2)

linear_error = metrics.mean_squared_error(y_test, linear_test)
random_error = metrics.mean_squared_error(y_test, random_test)
print(linear_error, random_error)

compare_table = pd.DataFrame()
compare_table["Vendas Reais"] = y_prod
compare_table["Previsao Random"] = random_test
compare_table = compare_table.reset_index(drop=True)

display(compare_table)
sns.lineplot(data=compare_table)
plt.show()

print(randomForest.feature_importances_)
print(ads_table[["Radio", "Jornal"]].sum())
```
